### Bivariate Analysis

```python
# Seperating dataframe into specific indicators
def create_dfs(df):
    indicators = [i for i in df['Indicator Name'].unique()]
    dfs = {}
    for i in indicators:
        dfs[i] = df[df['Indicator Name'] == i]
    return dfs

# Seperating new dataframe into now seperate countries
def create_country_dfs(df, periods=1):
    countries = [country for country in df['Country'].unique()]
    indicator_dfs = create_dfs(df) 
    country_dfs = {}
    
    for c in countries:
        country_dfs[c] = {} 
        for indicator, indicator_df in indicator_dfs.items():
            country_df = indicator_df[indicator_df['Country'] == c]
            country_df['Percentage Diff'] = country_df['Percentage'].diff(periods=periods)
            country_df['Percentage Diff'] = country_df['Percentage Diff'].fillna(0)
            country_dfs[c][indicator] = country_df
            
    return country_dfs
```

### Time Series Decomposition

```python
# Checking for stationarity
def adf_test_indicator(df, country):
    country_df = df[df['Country'] == country]
    indicator_results = {}
    
    unique_indicators = country_df['Indicator Name'].unique()
    for indicator in unique_indicators:
        indicator_data = country_df[country_df['Indicator Name'] == indicator]['Percentage']
        
        # Checking if the data is constant
        if indicator_data.var() == 0:
            continue  # Skipping the Dickey-Fuller test
        
        # Performing the Dickey-Fuller test
        result = adfuller(indicator_data)
        
        # Storing the result in the dictionary
        indicator_results[indicator] = {
            'ADF Statistic': result[0],
            'P-value': result[1],
            'Critical Values': result[4]
        }
    
    return indicator_results
```

```python
# List of countries for analysis
countries = ['Tanzania', 'Kenya', 'Uganda', 'Rwanda', 'Ethiopia', 'Sudan']

# Initializing an empty dictionary to store results
results = {}

# Iterating through each country in the list
for country in countries:
    # Calling the function adf_test_indicator to perform an Augmented Dickey-Fuller test
    country_results = adf_test_indicator(health_df, country)
    
    # Storing the results in the 'results' dictionary with the country name as the key
    results[country] = list(country_results.items())

# Printing the stationarity results for all countries
for country, country_results in results.items():
    print(f"Indicators in {country} and stationarity:")
    
    # Iterating through the indicators and their ADF test results
    for indicator, result in country_results:
        print(f"{indicator}:")
        
        # Checking if the ADF Statistic is less than the critical value at 1% significance level
        # and if the P-value is less than 0.05, indicating stationarity
        if result['ADF Statistic'] < result['Critical Values']['1%'] and result['P-value'] < 0.05:
            print("Stationary")
        else:
            print("Not Stationary")
        print()
```

Transforming time series data into a stationary form is a crucial step for effective application of time series models. This process enhances the accuracy and reliability of forecasting, simplifies model development, and aids in uncovering meaningful patterns and relationships within the data. The provided code substantiates the absence of stationarity for each indicator in each country, a critical observation validated by visualizations generated through the plot function.

To address the issue of non-stationarity and further refine the modeling process, dedicated efforts were made in the subsequent modeling section.
